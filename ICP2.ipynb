```python
def sum_all(args):
    if not args:
        return 0
    return sum_all(args[1:]) + args[0]

print(f'Sum of 1, 2, 3 is: {sum_all([1, 2, 3])}')
print(f'Sum of 1, 2, 3, 4, 5 is: {sum_all([1, 2, 3, 4, 5])}')
```

```python

def unique_list(lst):
    unique_items = []
    for item in lst:
        if item not in unique_items:
            unique_items.append(item)
    return unique_items

sample = [1, 2, 3, 3, 3, 3, 4, 5]
print("Unique List:", unique_list(sample))

```
