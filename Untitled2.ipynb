```python

import numpy as np

arr = np.random.randint(1, 21, 15)
print("Original Vector:\n", arr)

arr = arr.reshape(3, 5)
print("\nReshaped Array (3x5):\n", arr)

print("\nArray Shape:", arr.shape)

for i in range(arr.shape[0]):
    max_val = np.max(arr[i])
    arr[i][arr[i] == max_val] = 0

print("\nArray After Replacing Row Max with 0:\n", arr)
```

```python
import numpy as np

A = np.array([[3, -2],
              [1,  0]])

eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:")
print(eigenvalues)

print("Right eigenvectors (columns):")
print(eigenvectors)
```
